This notebook demonstrates how to use the Unravel Data Client to fetch OHLCV market data.

````python
import os
from datetime import date

from dotenv import load_dotenv

from unravel_data_client import get_metric, get_symbols

load_dotenv()
````

````python
API_KEY = os.getenv("UNRAVEL_API_KEY")
````

## List Available Symbols

Before fetching data, you can list all available symbols for an exchange.

````python
# List all available symbols for Binance Futures
symbols = get_symbols(
    api_key=API_KEY,
    exchange="binance-futures",
)

print(f"Found {len(symbols)} symbols for binance-futures")
print(f"\nFirst 20 symbols: {symbols[:20]}")
print(f"\nLooking for BTC pairs: {[s for s in symbols if 'btc' in s][:10]}")
````

## Fetch OHLCV Data for a Single Symbol

````python
df = get_metric(
    api_key=API_KEY,
    timestamp="true",
    interval="1h",
    exchange="binance-futures",
    symbol="btcusdt",
    start_date=date(2024, 1, 1),
    end_date=date(2024, 3, 31),
)
````

````python
df.head()
````

````python
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns}")
````

## Different Time Intervals

````python
# Fetch 1-minute data for a single day
df_1m = get_metric(
    api_key=API_KEY,
    timestamp="true",
    interval="1m",
    exchange="binance-futures",
    symbol="btcusdt",
    start_date=date(2024, 1, 15),
    end_date=date(2024, 1, 15),
)

print(f"1-minute bars for a single day: {df_1m.shape[0]} rows")
df_1m.head()
````
