## Fetch OHLCV Data for a Single Symbol

````python
df = get_ohlcv_historical(
    api_key=API_KEY,
    arrival_time="true",
    period="1h",
    exchange="binance-futures",
    symbol="btcusdt",
    start_date=date(2024, 1, 1),
    end_date=date(2024, 3, 31),
)
````

## Working with Polars

The library returns Polars DataFrames, which provide excellent performance for data analysis.

````python
import polars as pl

# Calculate daily returns using Polars
df_with_returns = df.with_columns(
    ((pl.col("close") - pl.col("close").shift(1)) / pl.col("close").shift(1) * 100)
    .alias("return_pct")
)

df_with_returns.select(["datetime", "close", "return_pct"]).head(10)
````

````python
# Filter and aggregate
daily_stats = (
    df_with_returns
    .filter(pl.col("return_pct").is_not_null())
    .select([
        pl.col("return_pct").mean().alias("avg_return"),
        pl.col("return_pct").std().alias("volatility"),
        pl.col("return_pct").min().alias("worst_return"),
        pl.col("return_pct").max().alias("best_return"),
        pl.col("volume").sum().alias("total_volume"),
    ])
)

daily_stats
````
